```python
import warnings
from contextlib import suppress

with suppress(Exception):
    warnings.filterwarnings("ignore")
```

```python
import pandas as pd

# CSV dosyasını oku
df_titanic = pd.read_csv("/kaggle/input/titanic-dataset-csv/Titanic_dataset.csv")

# İlk birkaç satıra bak
df_titanic.head()
```

### 1. Kazada ölenlerin yaş ortalaması

```python
df_titanic[df_titanic["Survived"] == 0]["Age"].mean()
```

### 2. Kazada ölenlerin bilet fiyatlarının ortalama ve medyanı


```python
dead = df_titanic[df_titanic["Survived"] == 0]
dead["Fare"].mean(), dead["Fare"].median()
```

### 3. Kazada ölen erkeklerin yaş ortalaması


```python
dead_men = df_titanic[(df_titanic["Survived"] == 0) & (df_titanic["Sex"] == "male")]
dead_men["Age"].mean()
```

### 4. Kazada ölen kadınların yaş ortalaması


```python
dead_women = df_titanic[(df_titanic["Survived"] == 0) & (df_titanic["Sex"] == "female")]
dead_women["Age"].mean()
```

### 5. Kazadan kurtulanların yaş ortalaması


```python
df_titanic[df_titanic["Survived"] == 1]["Age"].mean()
```

### 6. Kazadan kurtulanların bilet fiyatlarının ortalaması

```python
df_titanic[df_titanic["Survived"] == 1]["Fare"].mean()
```

### 7. Kazadan kurtulan toplam kişi sayısı


```python
len(df_titanic[df_titanic["Survived"] == 1])
```

### 8. 10 yaşından küçüklerin bilet fiyatlarının medyanı


```python
df_titanic[df_titanic["Age"] < 10]["Fare"].median()
```

### 9. Pclass'a göre bilet fiyatı istatistikleri


```python
df_titanic.groupby("Pclass")["Fare"].agg(["mean", "median"])
```

### 10. Kazada ölen kadın ve erkek oranı karşılaştırması


```python
male_dead_ratio = len(df_titanic[(df_titanic["Sex"] == "male") & (df_titanic["Survived"] == 0)]) / len(df_titanic[df_titanic["Sex"] == "male"])
female_dead_ratio = len(df_titanic[(df_titanic["Sex"] == "female") & (df_titanic["Survived"] == 0)]) / len(df_titanic[df_titanic["Sex"] == "female"])

male_dead_ratio, female_dead_ratio
```
